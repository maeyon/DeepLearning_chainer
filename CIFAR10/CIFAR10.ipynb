```python
import numpy as np
import json
import matplotlib.pyplot as plt
import pdb

data = np.load('SGLD/SGLDdata.npy')

n = 20
x = np.arange(n)
y_sum = np.zeros(n)
y = np.zeros(n)
sec = np.arange(1,n).astype(np.float32)
data[:,0] *= n

for i in xrange(5000):
    index = 0
    for s in sec:
        if data[i][0] < s:
            break
        else:
            index += 1
    y_sum[index] += 1.
    if data[i][1] == data[i][2]:
        y[index] += 1.

y = y / y_sum

plt.subplot(221)
plt.bar(x, y)
plt.title('SGLD entropy and accuracy')
plt.xlabel('entropy')
plt.ylabel('accuracy')
plt.ylim(0,1)

plt.subplot(222)
plt.hist(data[:,0], bins=20)
plt.title('SGLD entropy and freqency')
plt.xlabel('entropy')
plt.ylabel('freqency')

data = np.load('SGD/SGDdata.npy')

n = 20
x = np.arange(n)
y_sum = np.zeros(n)
y = np.zeros(n)
sec = np.arange(1,n).astype(np.float32)
data[:, 0] *= n

for i in xrange(5000):
    index = 0
    for s in sec:
        if data[i][0] < s:
            break
        else:
            index += 1
    y_sum[index] += 1.
    if data[i][1] == data[i][2]:
        y[index] += 1.

y = y / y_sum

plt.subplot(223)
plt.bar(x, y)
plt.title('SGD entropy and accuracy')
plt.xlabel('entropy')
plt.ylabel('accuracy')
plt.ylim(0,1)

plt.subplot(224)
plt.hist(data[:,0], bins=20)
plt.title('SGD entropy and freqency')
plt.xlabel('entropy')
plt.ylabel('freqency')

plt.tight_layout()

plt.show()
```
