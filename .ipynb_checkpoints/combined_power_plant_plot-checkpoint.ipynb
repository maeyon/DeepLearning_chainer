```python
%matplotlib notebook
#coding: utf-8
'''

result
'''
import pandas as pd
import xlrd
import numpy as np
with np.errstate(divide='ignore'):
    np.float64(1.0) / 0.0
import chainer
#from chainer import cuda
import chainer.functions as F
from chainer import optimizers
from chainer import initializer
from chainer.initializers import normal
import time
from sklearn.datasets import fetch_mldata
from sklearn.cross_validation import train_test_split
import pylab
import matplotlib.pyplot as plt
import scipy
from scipy.sparse import lil_matrix
import chainer.links as L

batchsize = 40
n_epoch = 2

#(0)
print("loading data...")
books = "Folds5x2_pp.xlsx"
book = xlrd.open_workbook(books)

sheet_1=book.sheet_by_name('Sheet1')
sheet_2=book.sheet_by_name('Sheet2')
sheet_3=book.sheet_by_name('Sheet3')
sheet_4=book.sheet_by_name('Sheet4')
sheet_5=book.sheet_by_name('Sheet5')

X=[]
Y=[]
 
for i in range(1,9569): # X[9567]
    X.append(sheet_1.row_values(i,start_colx=0,end_colx=4))
    Y.append(sheet_1.row_values(i,4))
for i in range(1,9569): 
    X.append(sheet_2.row_values(i,start_colx=0,end_colx=4))
    Y.append(sheet_2.row_values(i,4))
for i in range(1,9569): 
    X.append(sheet_3.row_values(i,start_colx=0,end_colx=4))
    Y.append(sheet_3.row_values(i,4))
for i in range(1,9569): 
    X.append(sheet_4.row_values(i,start_colx=0,end_colx=4))
    Y.append(sheet_4.row_values(i,4))
for i in range(1,9569): 
    X.append(sheet_5.row_values(i,start_colx=0,end_colx=4))
    Y.append(sheet_5.row_values(i,4))
X2=[]
Y2=[]
# print(X[9568],Y[9568])
for x in X:
    x = np.array(x)
    x = x.astype(np.float32)
    X2.append(x)
for y in Y:
    y = np.array(y)
    y = y.astype(np.float32)
    Y2.append(y)
X2=np.array(X2)
Y2=np.array(Y2)

X_train, X_test, Y_train, Y_test = train_test_split(X2, Y2, test_size=0.1)
X_train = X_train.reshape((len(X_train), 4))
X_test = X_test.reshape((len(X_test), 4))
Y_train=Y_train.reshape(-1,1)
Y_test=Y_test.reshape(-1,1)

#print(type(Y_test),Y_test.shape,Y_test.ndim)
#print(type(X_test),X_test.shape,X_test.ndim)

# (1)
model = chainer.FunctionSet(l1=L.Linear(4,64),#l1=semiLinearLayer(196,1000)
                            l3=L.Linear(64,1),
                            )


def forward(x_data, y_data, train=True,epoch=True):
    x, t = chainer.Variable(x_data), chainer.Variable(y_data)

    h1=F.sigmoid(model.l1(x))
    h2=model.l3(h1)
    y=h2
    if epoch:
        return F.mean_squared_error(y,t),y
    else:
        return F.mean_squared_error(y,t)
    
# (2)

optimizer = optimizers.Eve()
optimizer.setup(model)

# (3)
train_loss = []
test_loss = []
l_epoch=[]
train_mean_loss_=[]
test_mean_loss_=[]


N=Y_train.size
N_test=Y_test.size



for epoch in range(1, n_epoch + 1):
    print("epoch: %d" % epoch)

    perm = np.random.permutation(N)
    train_sum_loss = 0
    train_sum_acc=0
    for i in range(0, N, batchsize):
        x_batch = np.asarray(X_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])
        y_batch = np.asarray(Y_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])

        optimizer.zero_grads()
        loss,y = forward(x_batch, y_batch)
        loss.backward()
        optimizer.update()
        train_loss.append(loss.data)
        
        train_sum_loss += float(loss.data) * len(y_batch)
       
    l_epoch.append("epoch: %d" % epoch)
    train_mean_loss_.append(train_sum_loss/N)
    
    print("train mean loss={}".format(train_sum_loss / N))

    
    test_sum_accuracy = 0
    test_sum_loss=0
    for i in range(0, N_test, batchsize):
        x_batch = np.asarray(X_test[i:(i + batchsize) if (i+batchsize) < N else N ])
        y_batch = np.asarray(Y_test[i:(i + batchsize) if (i+batchsize) < N else N ])
    
        if epoch==1000:
            loss,y = forward(x_batch, y_batch, train=False)
            print(y.data)
            
        else:
            loss = forward(x_batch, y_batch, train=False,epoch=False)
        test_loss.append(loss.data)
        test_sum_loss += float(loss.data) * len(y_batch)
        
        
    print("test mean loss={}".format(test_sum_loss/N_test))
    test_mean_loss_.append(test_sum_loss/N_test)

'''
# 
'''


model2 = chainer.FunctionSet(l1=L.Linear(4,128),
                            l3=L.Linear(128,1),
                            )


def forward2(x_data, y_data, train=True,epoch=True):
    x, t = chainer.Variable(x_data), chainer.Variable(y_data)

    h1=F.sigmoid(model2.l1(x))
    h3=model2.l3(h1)
    y=h3
    if epoch:
        return F.mean_squared_error(y,t),y
    else:
        return F.mean_squared_error(y,t)
    
# (2)

optimizer2 = optimizers.Adam()
optimizer2.setup(model2)

# (3)
train_loss = []
test_loss = []
l_epoch=[]
train_mean_loss_2=[]
test_mean_loss_2=[]


N=Y_train.size
N_test=Y_test.size



for epoch in range(1, n_epoch + 1):
    print("epoch: %d" % epoch)

    perm = np.random.permutation(N)
    train_sum_loss= 0
    train_sum_acc=0
    for i in range(0, N, batchsize):
        x_batch = np.asarray(X_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])
        y_batch = np.asarray(Y_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])

        optimizer2.zero_grads()
        loss,y = forward2(x_batch, y_batch)
        loss.backward()
        optimizer2.update()
        train_loss.append(loss.data)
        
        train_sum_loss += float(loss.data) * len(y_batch)
       
    l_epoch.append("epoch: %d" % epoch)
    train_mean_loss_2.append(train_sum_loss/N)
    
    print("train mean loss={}".format(train_sum_loss / N))
    
    test_sum_accuracy = 0
    test_sum_loss=0
    
    for i in range(0, N_test, batchsize):
        x_batch = np.asarray(X_test[i:(i + batchsize) if (i+batchsize) < N else N ])
        y_batch = np.asarray(Y_test[i:(i + batchsize) if (i+batchsize) < N else N ])
    
        if epoch==1000:
            loss,y = forward2(x_batch, y_batch, train=False)
            print(y.data)
            
        else:
            loss = forward2(x_batch, y_batch, train=False,epoch=False)
        test_loss.append(loss.data)
        test_sum_loss += float(loss.data) * len(y_batch)
        
        
    print("test mean loss={}".format(test_sum_loss/N_test))
    test_mean_loss_2.append(test_sum_loss/N_test)


'''

'''


model3 = chainer.FunctionSet(l1=L.Linear(4,256),
                            l4=L.Linear(256,1)
                            )


def forward3(x_data, y_data, train=True,epoch=True):
    x, t = chainer.Variable(x_data), chainer.Variable(y_data)

    h1=F.sigmoid(model3.l1(x))
    h4=(model3.l4(h1))
    y=h4
    
    if epoch:
        return F.mean_squared_error(y,t),y
    else:
        return F.mean_squared_error(y,t)
    
# (2)

optimizer3 = optimizers.Adam()
optimizer3.setup(model3)

# (3)
train_loss = []
test_loss = []
l_epoch=[]
train_mean_loss_3=[]
test_mean_loss_3=[]


N=Y_train.size
N_test=Y_test.size

for epoch in range(1, n_epoch + 1):
    print("epoch: %d" % epoch)

    perm = np.random.permutation(N)
    train_sum_loss= 0
    train_sum_acc=0
    for i in range(0, N, batchsize):
        x_batch = np.asarray(X_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])
        y_batch = np.asarray(Y_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])

        optimizer3.zero_grads()
        loss,y = forward3(x_batch, y_batch)
        loss.backward()
        optimizer3.update()
        train_loss.append(loss.data)
        
        train_sum_loss += float(loss.data) * len(y_batch)
       
    l_epoch.append("epoch: %d" % epoch)
    train_mean_loss_3.append(train_sum_loss/N)
    
    print("train mean loss={}".format(train_sum_loss / N))

    test_sum_accuracy = 0
    test_sum_loss=0
    #l3_W=[]
    for i in range(0, N_test, batchsize):
        x_batch = np.asarray(X_test[i:(i + batchsize) if (i+batchsize) < N else N ])
        y_batch = np.asarray(Y_test[i:(i + batchsize) if (i+batchsize) < N else N ])
    
        if epoch==1000:
            loss,y = forward3(x_batch, y_batch, train=False)
            print(y.data)
            
        else:
            loss = forward3(x_batch, y_batch, train=False,epoch=False)
        test_loss.append(loss.data)
        test_sum_loss += float(loss.data) * len(y_batch)
        
        
    print("test mean loss={}".format(test_sum_loss/N_test))
    test_mean_loss_3.append(test_sum_loss/N_test)
    
'''

'''


model4 = chainer.FunctionSet(l1=L.Linear(4,128),
                            l3=L.Linear(128,1)
                            )


def forward4(x_data, y_data, train=True,epoch=True):
    x, t = chainer.Variable(x_data), chainer.Variable(y_data)

    h1=F.relu(model4.l1(x))
    h3=model4.l3(h1)
    y=h3
    #print(y.ndim,y.shape)
    #print(t.ndim,t.shape)
    if epoch:
        return F.mean_squared_error(y,t),y
    else:
        return F.mean_squared_error(y,t)
    
# (2)

optimizer4 = optimizers.Adam()
optimizer4.setup(model4)

# (3)
train_loss = []
test_loss = []
l_epoch=[]
train_mean_loss_4=[]
test_mean_loss_4=[]

N=Y_train.size
N_test=Y_test.size

for epoch in range(1, n_epoch + 1):
    print("epoch: %d" % epoch)

    perm = np.random.permutation(N)
    train_sum_loss= 0
    train_sum_acc=0
    for i in range(0, N, batchsize):
        x_batch = np.asarray(X_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])
        y_batch = np.asarray(Y_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])

        optimizer4.zero_grads()
        loss,y = forward4(x_batch, y_batch)
        loss.backward()
        optimizer4.update()
        train_loss.append(loss.data)
        
        train_sum_loss += float(loss.data) * len(y_batch)
       
    l_epoch.append("epoch: %d" % epoch)
    train_mean_loss_4.append(train_sum_loss/N)
    
    print("train mean loss={}".format(train_sum_loss / N))

    test_sum_accuracy = 0
    test_sum_loss=0
    for i in range(0, N_test, batchsize):
        x_batch = np.asarray(X_test[i:(i + batchsize) if (i+batchsize) < N else N ])
        y_batch = np.asarray(Y_test[i:(i + batchsize) if (i+batchsize) < N else N ])
    
        if epoch==1000:
            loss,y = forward4(x_batch, y_batch, train=False)
            print(y.data)
            
        else:
            loss = forward4(x_batch, y_batch, train=False,epoch=False)
        test_loss.append(loss.data)
        test_sum_loss += float(loss.data) * len(y_batch)
        
        
    print("test mean loss={}".format(test_sum_loss/N_test))
    test_mean_loss_4.append(test_sum_loss/N_test)
    

'''

'''


model5 = chainer.FunctionSet(l1=L.Linear(4,256),
                            l2=L.Linear(256,1)
                            )


def forward5(x_data, y_data, train=True,epoch=True):
    x, t = chainer.Variable(x_data), chainer.Variable(y_data)

    h1=F.relu(model5.l1(x))
    h2=model5.l2(h1)
    y=h2
    
    if epoch:
        return F.mean_squared_error(y,t),y
    else:
        return F.mean_squared_error(y,t)
    
# (2)

optimizer5 = optimizers.Adam()
optimizer5.setup(model5)

# (3)
train_loss = []
test_loss = []
l_epoch=[]
train_mean_loss_5=[]
test_mean_loss_5=[]

N=Y_train.size
N_test=Y_test.size

for epoch in range(1, n_epoch + 1):
    print("epoch: %d" % epoch)

    perm = np.random.permutation(N)
    train_sum_loss= 0
    train_sum_acc=0
    for i in range(0, N, batchsize):
        x_batch = np.asarray(X_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])
        y_batch = np.asarray(Y_train[perm[i:(i + batchsize) if (i+batchsize) < N else N ]])

        optimizer5.zero_grads()
        loss,y = forward5(x_batch, y_batch)
        loss.backward()
        optimizer5.update()
        train_loss.append(loss.data)
        
        train_sum_loss += float(loss.data) * len(y_batch)
       
    l_epoch.append("epoch: %d" % epoch)
    train_mean_loss_5.append(train_sum_loss/N)
    
    print("train mean loss={}".format(train_sum_loss / N))

    test_sum_accuracy = 0
    test_sum_loss=0
    for i in range(0, N_test, batchsize):
        x_batch = np.asarray(X_test[i:(i + batchsize) if (i+batchsize) < N else N ])
        y_batch = np.asarray(Y_test[i:(i + batchsize) if (i+batchsize) < N else N ])
    
        if epoch==1000:
            loss,y = forward5(x_batch, y_batch, train=False)
            print(y.data)
            
        else:
            loss = forward5(x_batch, y_batch, train=False,epoch=False)
        test_loss.append(loss.data)
        test_sum_loss += float(loss.data) * len(y_batch)
        
        
    print("test mean loss={}".format(test_sum_loss/N_test))
    test_mean_loss_5.append(test_sum_loss/N_test)



# (Z)
plt.figure(figsize=(20,10))

#plt.legend(["train_loss","test_loss"],loc=4)



plt.subplot(2,2,1)
plt.plot(range(1,len(l_epoch)+1), train_mean_loss_,'r.-',label='model1')
plt.plot(range(1,len(l_epoch)+1), train_mean_loss_2,'g.-',label='model2')
plt.plot(range(1,len(l_epoch)+1), train_mean_loss_3,'b.-',label='model3')
plt.plot(range(1,len(l_epoch)+1), train_mean_loss_4,'y.-',label='model4')
plt.plot(range(1,len(l_epoch)+1), train_mean_loss_5,'k.-',label='model5')
#plt.hlines([0.1,0.45],0,len(l_epoch),linestyles="dashed")
plt.title("Loss of train data")
plt.xlabel("epoch",fontname='serif')
plt.legend(loc=1)

plt.subplot(2,2,2)
#plt.hlines([0.98,0.85],0,len(l_epoch),linestyles="dashed")
plt.plot(range(1,len(l_epoch)+1), test_mean_loss_,'r.-',label='model1')
plt.plot(range(1,len(l_epoch)+1), test_mean_loss_2,'g.-',label='model2')
plt.plot(range(1,len(l_epoch)+1), test_mean_loss_3,'b.-',label='model3')
plt.plot(range(1,len(l_epoch)+1), test_mean_loss_4,'y.-',label='model4')
plt.plot(range(1,len(l_epoch)+1), test_mean_loss_5,'k.-',label='model5')
plt.title('Loss of test data')
plt.xlabel("epoch",fontname='serif') 
plt.legend(loc=1)

plt.subplot(2,2,4)
#plt.hlines([0.98,0.85],0,len(l_epoch),linestyles="dashed")
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), test_mean_loss_[int(len(l_epoch)/2):],'r.-',label='model1')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), test_mean_loss_2[int(len(l_epoch)/2):],'g.-',label='model2')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), test_mean_loss_3[int(len(l_epoch)/2):],'b.-',label='model3')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), test_mean_loss_4[int(len(l_epoch)/2):],'y.-',label='model4')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), test_mean_loss_5[int(len(l_epoch)/2):],'k.-',label='model5')
plt.title('Loss of test data(posterior half)')
plt.xlabel("epoch",fontname='serif') 
plt.legend(loc=1)

plt.subplot(2,2,3)
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), train_mean_loss_[int(len(l_epoch)/2):],'r.-',label='model1')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), train_mean_loss_2[int(len(l_epoch)/2):],'g.-',label='model2')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), train_mean_loss_3[int(len(l_epoch)/2):],'b.-',label='model3')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), train_mean_loss_4[int(len(l_epoch)/2):],'y.-',label='mode14')
plt.plot(range(int(len(l_epoch)/2)+1,len(l_epoch)+1), train_mean_loss_5[int(len(l_epoch)/2):],'k.-',label='mode15')
#plt.hlines([0.1,0.45],0,len(l_epoch),linestyles="dashed")
#plt.xticks(l_epoch[int(len(l_epoch)/2):])
plt.title("Loss of train data(posterior half)")
plt.xlabel("epoch",fontname='serif')
plt.legend(loc=1)



plt.show()
```
