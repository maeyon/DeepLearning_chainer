```python
import numpy as np
import json
import matplotlib.pyplot as plt
import pdb

data = np.loadtxt('SGLD/SGLDdata.csv', delimiter=',')

n = 20
x = np.arange(n)
y_sum = np.zeros(n)
y = np.zeros(n)
sec = np.arange(1,n).astype(np.float32) / n * data[:, 0].max()

for i in xrange(len(data)):
    index = 0
    for s in sec:
        if data[i][0] < s:
            break
        else:
            index += 1
    y_sum[index] += 1.
    if data[i][1] == data[i][2]:
        y[index] += 1.

y = y / y_sum

plt.subplot(121)
plt.bar(x, y)
plt.title('SGLD entropy and accuracy')
plt.xlabel('entropy')
plt.ylabel('accuracy')

p = np.load('SGLDaft/bayesian.npy').astype(np.float32)
p /= p.sum(axis=1, keepdims=True)
predict = p.argmax(axis=1)
p[p < 1e-30] = 1e-30
entropy = -np.sum(p * np.log(p), axis=1)

n = 20
x = np.arange(n)
y_sum = np.zeros(n)
y = np.zeros(n)
sec = np.arange(1,n).astype(np.float32) / n * entropy.max()

for i in xrange(len(entropy)):
    index = 0
    for s in sec:
        if entropy[i] < s:
            break
        else:
            index += 1
    y_sum[index] += 1.
    if predict[i] == data[i][2]:
        y[index] += 1.

y = y / y_sum

plt.subplot(122)
plt.bar(x, y)
plt.title('HEL entropy and accuracy')
plt.xlabel('entropy')
plt.ylabel('accuracy')

plt.tight_layout()

plt.show()
```
