```python
import numpy as np
import matplotlib.pyplot as plt

image = np.load('test.npy')
data = np.loadtxt('SGLDdata.csv', delimiter=',')

i0 = []
i1 = []
for i in xrange(1000):
    if data[i][1] < 0.5:
        i0.append(i)
    else:
        i1.append(i)

ih = []
im = []
il = []
for i in xrange(1000):
    if data[i][0] < 1e-3:
        il.append(i)
    elif data[i][0] < 1e-1:
        im.append(i)
    else:
        ih.append(i)

plt.subplot(131)
plt.plot(image[:500, 0], image[:500, 1], 'r.', label='class 0')
plt.plot(image[500:, 0], image[500:, 1], 'b.', label='class 1')
plt.legend(bbox_to_anchor=(1, 0))
plt.title('image')

plt.subplot(132)
plt.plot(image[i0, 0], image[i0, 1], 'r.', label='class 0')
plt.plot(image[i1, 0], image[i1, 1], 'b.', label='class 1')
plt.legend(bbox_to_anchor=(1, 0))
plt.title('predict')

plt.subplot(133)
plt.plot(image[il, 0], image[il, 1], 'b.', label='low')
plt.plot(image[im, 0], image[im, 1], 'y.', label='medium')
plt.plot(image[ih, 0], image[ih, 1], 'r.', label='high')
plt.legend(bbox_to_anchor=(1, 0))
plt.title('entropy')

plt.tight_layout()

plt.show()
```
